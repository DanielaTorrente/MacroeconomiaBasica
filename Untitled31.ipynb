# üìò Modelo Keynesiano Simple ‚Äì Equilibrio Macroecon√≥mico con Gobierno

Este modelo analiza el nivel de producci√≥n y empleo en el corto plazo bajo los supuestos de precios constantes y existencia de recursos ociosos. La producci√≥n est√° determinada por la **Demanda Global efectiva de periodo (Z)**.

---

## üîç Supuestos del modelo

- Econom√≠a **cerrada**: sin exportaciones ni importaciones.
- Tres sectores institucionales: **familias**, **empresas** y **gobierno**.
- **Precios constantes**: Los  precios no se modifican aunque varien las cantidades. esto se produce debido a la presencia de desempleo involuntario.
- La **Oferta Global (Y)** est√° determinada por la **Demanda Global (Z)**.
- Las empresas pueden producir cualquier cantidad que se demande (oferta perfectamente el√°stica).
- El equilibrio se representa gr√°ficamente por la intersecci√≥n entre **Z** y la bisectriz **Y = Z**.

---

## üßæ Componentes de la Demanda Global

La **Demanda Global (Z)** est√° compuesta por:

$$
Z = C + I + G
$$

Donde:

- \( C = Co + c(Y - T) \): funci√≥n de consumo en funci√≥n del ingreso disponible.
- \( I = Io \): inversi√≥n aut√≥noma de las empresas.
- \( G = Go \): gasto p√∫blico.
- \( To \): impuestos aut√≥nomos.

Sustituyendo todos los componentes:

$$
Z = C_0 + c(Y - T_0) + I_0 + G_0
$$

---

## ‚öñÔ∏è Condici√≥n de Equilibrio

El equilibrio macroecon√≥mico se alcanza cuando:

$$
Y = Z
$$

Es decir, la **Oferta Global (Y)** es igual a la **Demanda Global  (Z)**.

---

## üßÆ Ingreso de Equilibrio

Resolviendo la ecuaci√≥n de equilibrio:

$$
Y = \dfrac{C_0 - cT_0 + I_0 + G_0}{1 - c}
$$

---

### üß© Interpretaci√≥n

- El **numerador** representa el **componente aut√≥nomo de la Demanda Global**:

$$
A = C_0 - cT_0 + I_0 + G_0
$$

- El **denominador** representa el **inverso del multiplicador keynesiano**:

$$
k = \dfrac{1}{1 - c}
$$

Entonces:

$$
Y = k \cdot A
$$

---
### ‚ÑπÔ∏è La propensi√≥n marginal a consumir

La propensi√≥n marginal a consumir es la proporci√≥n del ingreso disponible que se destina al consumo cuando la renta disponible var√≠a en una unidad. Por lo tanto, refleja cu√°nto se incrementa el consumo si aumenta el ingreso.

Este par√°metro se encuentra entre 0 y 1, y suele considerarse cercano a 1 en econom√≠as con alta propensi√≥n al consumo. Cuanto mayor es su valor, m√°s sensible es la demanda ante variaciones en el ingreso.

La estabilidad de esta relaci√≥n es lo que permite que funcione el mecanismo del multiplicador: un aumento inicial en el gasto genera una cadena de incrementos sucesivos en el ingreso y el consumo, que se retroalimentan.
## üìà Representaci√≥n Gr√°fica del Equilibrio

- La curva **Z** se representa como una l√≠nea recta con pendiente \( c \) y ordenada al origen igual a \( A \).
- La l√≠nea **Y = Z** es la bisectriz: representa todos los puntos donde la producci√≥n es igual a la demanda.
- El equilibrio se da en la intersecci√≥n entre la curva **Z** (Demanda Global) y la bisectriz **Y** (Oferta Global).

---


```python
# @title
import matplotlib.pyplot as plt
import numpy as np

# Crear valores para el ingreso Y
Y = np.linspace(0, 3000, 100)

# Par√°metros del modelo
C0 = 400           # Consumo aut√≥nomo
c = 0.5            # Propensi√≥n marginal a consumir
T0 = 200           # Impuestos aut√≥nomos
I0 = 200           # Inversi√≥n aut√≥noma
G0 = 500           # Gasto p√∫blico aut√≥nomo

# Demanda Global Z
Z = C0 + c * (Y - T0) + I0 + G0

# Componente aut√≥nomo A
A = C0 - c * T0 + I0 + G0

# Ingreso de equilibrio Y*
Y_eq = A / (1 - c)

# Gr√°fico
plt.figure(figsize=(10, 6))

# Curva de Demanda Global Z
plt.plot(Y, Z, label=r'$Z = C_0 + c(Y - T_0) + I_0 + G_0$', linewidth=2)

# Oferta Global Y (bisectriz)
plt.plot(Y, Y, '--', color='black', label=r'$Y = Z$ (Bisectriz)')

# L√≠nea horizontal para el componente aut√≥nomo A
plt.axhline(y=A, color='gray', linestyle=':', linewidth=1)
plt.text(50, A + 30, r'$A = C_0 - cT_0 + I_0 + G_0$', fontsize=11, color='gray')

# L√≠nea punteada vertical en el equilibrio Y*
plt.axvline(x=Y_eq, color='red', linestyle=':', linewidth=1)
plt.text(Y_eq + 20, 100, r'$Y^*$', color='red', fontsize=12)

# Est√©tica
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.xticks([])  # Sin valores num√©ricos en eje X
plt.yticks([])  # Sin valores num√©ricos en eje Y

plt.title("Modelo Keynesiano Simple ‚Äì Equilibrio entre Z y Y", fontsize=14)
plt.xlabel(r'Oferta Global $(Y)$')
plt.ylabel(r'Demanda Global $(Z)$')
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)

# Ejes
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)

plt.tight_layout()
plt.show()
```

## üß≠ Cambios en el Equilibrio por Variaciones en la Demanda Global

El ingreso de equilibrio en el modelo keynesiano simple con gobierno est√° dado por:

$$
Y = \dfrac{C_0 - cT_0 + I_0 + G_0}{1 - c}
$$

Donde:
- \( Co \): consumo aut√≥nomo
- \( Io \): inversi√≥n aut√≥noma
- \( Go \): gasto p√∫blico
- \( To \): impuestos aut√≥nomos
- \( c \): propensi√≥n marginal a consumir

---

### üéØ Multiplicador keynesiano

Se define como:

$$
k = \dfrac{1}{1 - c}
$$

Este valor indica cu√°ntas unidades se incrementa el ingreso total \( Y \) ante una variaci√≥n de una unidad monetaria en alg√∫n componente aut√≥nomo de la demanda.

-

El efecto es **directo inicialmente** y se amplifica por el multiplicador que opera a partir del los sucesivos incrementos del consumo generados por los aumantos de la renta.

## üîÅ Mecanismo multiplicador

Cuando alguno de los componentes aut√≥nomos de la Demanda Global, como el gasto p√∫blico \( G_0 \), aumenta:

1. Se genera un **aumento inicial en la producci√≥n**, producto del mayor gasto directo en bienes y servicios.
2. Este aumento en el ingreso **eleva el ingreso disponible de las familias**, lo cual incrementa el consumo en una proporci√≥n determinada por \( c \), la propensi√≥n marginal a consumir.
3. Ese nuevo gasto en consumo **incrementa nuevamente la producci√≥n**, generando un nuevo ingreso, que a su vez vuelve a inducir consumo, y as√≠ sucesivamente.
4. Este proceso **se repite en valores cada vez menores**, ya que la propensi√≥n marginal a consumir es positiva pero menor que uno.

---

### üìê Resultado: progresi√≥n geom√©trica

El efecto total del incremento del gasto se comporta como una **progresi√≥n geom√©trica** de raz√≥n \( c \):

$$
\text{Impacto total en } Y = \text{impacto inicial} \cdot \left(1 + c + c^2 + c^3 + \cdots\right)
$$

Esta serie converge y su suma es:

$$
\frac{1}{1 - c}
$$

que es justamente el **multiplicador keynesiano**.

---

‚úÖ Por eso, **una pol√≠tica fiscal expansiva**, como un aumento en el gasto p√∫blico, genera un efecto total mayor que el inicial, **siempre que exista capacidad productiva ociosa** en la econom√≠a.


## üìò Multiplicador de los impuestos aut√≥nomos

Cuando el gobierno incrementa el nivel de impuestos aut√≥nomos, representado por T‚ÇÄ, se produce una disminuci√≥n en el ingreso disponible de las familias. Como el consumo depende del ingreso disponible, el gasto de consumo cae. Esta reducci√≥n en el consumo genera una ca√≠da inicial en la demanda global.

A su vez, esa ca√≠da en la demanda reduce la producci√≥n y el ingreso, lo que provoca una nueva reducci√≥n del ingreso disponible y, con ello, una nueva disminuci√≥n del consumo. Este mecanismo se repite sucesivamente, generando efectos cada vez menores, ya que la propensi√≥n marginal a consumir es un n√∫mero positivo pero menor que uno.

---

### üîÅ Efecto multiplicador de los impuestos aut√≥nomos

Este proceso tambi√©n se puede representar como una progresi√≥n geom√©trica, donde cada efecto sucesivo es proporcional al anterior. En este caso, como el efecto inicial es una ca√≠da del consumo, el resultado total es una disminuci√≥n del ingreso de equilibrio.

El valor del multiplicador de los impuestos aut√≥nomos est√° dado por:

$$
k_T = -\frac{c}{1 - c}
$$

---

### ‚úèÔ∏è Interpretaci√≥n

- El signo negativo indica que un aumento en los impuestos aut√≥nomos reduce el ingreso de equilibrio.
- La magnitud de este multiplicador es menor que la del gasto, ya que los impuestos afectan al ingreso disponible, y solo una fracci√≥n de esa variaci√≥n se traduce en consumo.
- Cuanto mayor sea la propensi√≥n marginal a consumir, mayor ser√° el impacto negativo de los impuestos sobre el ingreso.

---

### üìä Comparaci√≥n con el multiplicador del gasto



---

Por lo tanto, una pol√≠tica fiscal expansiva basada en una reducci√≥n de impuestos aut√≥nomos genera un aumento del ingreso de equilibrio, aunque su efecto total ser√° menor que el de un aumento equivalente del gasto p√∫blico.

### üìä Comparaci√≥n con el multiplicador del gasto

| Componente afectado        | Multiplicador          | Efecto sobre el ingreso |
|----------------------------|------------------------|--------------------------|
| Gasto p√∫blico              | 1 / (1 - c)            | Aumenta                 |
| Impuestos aut√≥nomos        | -c / (1 - c)           | Disminuye               |



## üìò La pol√≠tica fiscal en la teor√≠a keynesiana

En la teor√≠a keynesiana, la pol√≠tica fiscal es una herramienta fundamental para estabilizar la econom√≠a. Se basa en la intervenci√≥n activa del Estado a trav√©s del gasto p√∫blico y la recaudaci√≥n impositiva para influir sobre la demanda global y, por ende, sobre el nivel de producci√≥n y empleo.

Desde esta perspectiva, los mercados no tienden autom√°ticamente al pleno empleo. Por eso, cuando la econom√≠a enfrenta una ca√≠da en la demanda global, el Estado puede contrarrestarla mediante el aumento del gasto p√∫blico o la reducci√≥n de impuestos, generando un efecto multiplicador que incrementa el ingreso y el consumo.

La pol√≠tica fiscal tiene dos componentes principales:

- **Gasto p√∫blico:** es una entrada directa en el flujo de la demanda. Su incremento puede generar un aumento sostenido del ingreso a trav√©s del mecanismo del multiplicador.
- **Impuestos:** afectan negativamente al ingreso disponible de las familias y, por lo tanto, al consumo. Reducirlos puede estimular la demanda, mientras que aumentarlos puede moderarla.

En el modelo keynesiano, la efectividad de la pol√≠tica fiscal depende del valor de la propensi√≥n marginal a consumir. Cuanto m√°s alta sea, mayor ser√° el impacto del gasto p√∫blico o de una reducci√≥n impositiva sobre el ingreso de equilibrio.

Supongamos los siguientes valores:

- \( Co = 400 \)

- \( To = 200 \)  
- \( c = 0,5 \) (propensi√≥n marginal a consumir)

**Objetivo:**

2. Simular un aumento individual en:
   - \( Co \) a 500
   - \( Io \) a 300
   - \( Go \) a 600
3. Calcular el nuevo \( Y* \) en cada caso.

5. Verificar si el cambio en \( Y \) coincide con la predicci√≥n del multiplicador.

---


```python
# @title
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Funci√≥n general para estimar ingreso de equilibrio, multiplicador y gasto aut√≥nomo
def estimar_equilibrio(Co=400, Io=200, Go=500, To=200, c=0.5, graficar=True):
    # Gasto aut√≥nomo
    A = Co + Io + Go - c * To
    # Multiplicador
    k = 1 / (1 - c)
    # Ingreso de equilibrio
    Y_eq = A * k
    #Multiplicador de To
    kT = -c / (1 - c)


    # Mostrar resultados
    resultados = {

        "Gasto aut√≥nomo (A)": A,
        "Multiplicador (k)": k,
        "Ingreso de equilibrio (Y*)": Y_eq,
        "Multiplicador de T": kT

    }

    df_resultado = pd.DataFrame.from_dict(resultados, orient='index', columns=['Valor'])


    # Gr√°fico opcional
    if graficar:
        Ys = np.linspace(0, Y_eq + 300, 100)
        Z = A + c * Ys
        plt.figure(figsize=(8,5))
        plt.plot(Ys, Ys, '--', label='L√≠nea 45¬∞', color='gray')
        plt.plot(Ys, Z, label='Curva Z (DA)')
        plt.scatter([Y_eq], [Y_eq], color='red', zorder=5)
        plt.text(Y_eq + 10, Y_eq - 30, f'Y* = {Y_eq:.1f}', color='red')
        plt.xlabel('Ingreso (Y)')
        plt.ylabel('Demanda Agregada (Z)')
        plt.title('Equilibrio Macroecon√≥mico')
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()

    return df_resultado

# Ejemplo de uso con valores por defecto
estimar_equilibrio()
```

```python
# @title
# ‚öôÔ∏è Simulador interactivo con bot√≥n de reinicio, flecha ŒîY y ordenada A

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output

# Widgets de entrada
Co_w = widgets.IntSlider(value=400, min=0, max=1000, step=50, description='Co')
Io_w = widgets.IntSlider(value=200, min=0, max=1000, step=50, description='Io')
Go_w = widgets.IntSlider(value=500, min=0, max=1000, step=50, description='Go')
To_w = widgets.IntSlider(value=200, min=0, max=1000, step=50, description='To')
c_w = widgets.FloatSlider(value=0.5, min=0.1, max=0.9, step=0.05, description='c')

# Bot√≥n de reinicio
reset_btn = widgets.Button(description="üîÅ Reiniciar valores")
out = widgets.Output()

# Funci√≥n gr√°fica principal
def actualizar_grafico(change=None):
    with out:
        clear_output(wait=True)

        # Par√°metros base (para comparaci√≥n)
        Co_base = 400
        Io_base = 200
        Go_base = 500
        To_base = 200
        c_base = 0.5

        # Nuevos par√°metros (usuario)
        Co = Co_w.value
        Io = Io_w.value
        Go = Go_w.value
        To = To_w.value
        c = c_w.value

        # C√°lculos base y nuevos
        A_base = Co_base + Io_base + Go_base - c_base * To_base
        k_base = 1 / (1 - c_base)
        Y_base = A_base * k_base

        A = Co + Io + Go - c * To
        k = 1 / (1 - c)
        Y_eq = A * k
        kT = -c /(1 - c)

        resultados = {
            "Gasto aut√≥nomo (A)": A,
            "Multiplicador (k)": k,
            "Ingreso de equilibrio (Y*)": Y_eq,
            "Multiplicador de T (kT)" :kT
            }
        df_resultado = pd.DataFrame.from_dict(resultados, orient='index', columns=['Valor'])

        Ys = np.linspace(0, max(Y_base, Y_eq) + 300, 100)
        Z_base = A_base + c_base * Ys
        Z = A + c * Ys

        plt.figure(figsize=(9,6))
        plt.plot(Ys, Ys, '--', label='L√≠nea 45¬∞', color='gray')
        plt.plot(Ys, Z_base, label='Z Base', color='blue')
        plt.plot(Ys, Z, label='Z Nueva', color='green')
        plt.scatter([Y_base], [Y_base], color='blue', zorder=5)
        plt.text(Y_base + 10, Y_base - 30, f'Y* Base = {Y_base:.1f}', color='blue')
        plt.scatter([Y_eq], [Y_eq], color='green', zorder=5)
        plt.text(Y_eq + 10, Y_eq - 30, f'Y* Nueva = {Y_eq:.1f}', color='green')

        # Flecha ŒîY
        plt.annotate('', xy=(Y_eq, 0), xytext=(Y_base, 0),
                     arrowprops=dict(arrowstyle='<|-|>', color='crimson', lw=2))
        plt.text((Y_base + Y_eq)/2, 20, f'ŒîY = {Y_eq - Y_base:.1f}',
                 ha='center', va='bottom', fontsize=10, color='crimson')

        # Ordenada A
        plt.scatter([0], [A], color='purple', zorder=5)
        plt.text(10, A + 10, f'A = {A:.1f}', color='purple')

        plt.xlabel('Ingreso (Y)')
        plt.ylabel('Demanda Agregada (Z)')
        plt.title('Modelo keynesiano: equilibrio y desplazamientos')
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()

        display(df_resultado)

# Conectar controles al gr√°fico
Co_w.observe(actualizar_grafico, names='value')
Io_w.observe(actualizar_grafico, names='value')
Go_w.observe(actualizar_grafico, names='value')
To_w.observe(actualizar_grafico, names='value')
c_w.observe(actualizar_grafico, names='value')

# Acci√≥n del bot√≥n de reinicio
def reiniciar(val):
    Co_w.value = 400
    Io_w.value = 200
    Go_w.value = 500
    To_w.value = 200
    c_w.value = 0.5

reset_btn.on_click(reiniciar)

# Mostrar todo
display(widgets.HBox([Co_w, Io_w, Go_w, To_w, c_w, reset_btn]))
display(out)

# Mostrar gr√°fico inicial
actualizar_grafico()
```




### üìÑ Enfoque de gastos compensatorios y filtraciones

En el modelo keynesiano de determinaci√≥n de la renta, es posible analizar el equilibrio macroecon√≥mico desde una perspectiva alternativa a la de la demanda agregada: el enfoque de **gastos compensatorios** y **filtraciones**.

Este m√©todo se basa en identificar:

- Por un lado, las **entradas al flujo de la renta** que **no provienen directamente de la producci√≥n**, pero que permiten financiar pagos y mantener el nivel de gasto. Estas son las **entradas o gastos compensatorios**, como:
  - La inversi√≥n (\( I \))
  - El gasto p√∫blico (\( G \))
  - Las transferencias (\( Tr \)) si estuvieran presentes

- Por otro lado, los **usos del ingreso** que **no se destinan a la compra de bienes y servicios nacionales**, es decir, las **filtraciones**, que incluyen:
  - El ahorro (\( S \))
  - Los impuestos (\( T \))

El equilibrio se alcanza cuando ambos flujos se igualan:

###       Entradas compensatorias = Filtraciones

Este enfoque permite entender el rol activo del Estado en la determinaci√≥n del nivel de actividad econ√≥mica y facilita el an√°lisis de los efectos de la pol√≠tica fiscal. Es particularmente √∫til para identificar cu√°les son los componentes que dinamizan el gasto y cu√°les act√∫an como freno al mismo.

### üìÑ Introducci√≥n al enfoque de gastos compensatorios y filtraciones

En el modelo keynesiano de determinaci√≥n de la renta, es posible analizar el equilibrio macroecon√≥mico desde una perspectiva alternativa a la de la demanda agregada: el enfoque de **gastos compensatorios** y **filtraciones**.

Este m√©todo se basa en identificar:

* Por un lado, las **entradas al flujo de la renta** que **no provienen directamente de la producci√≥n**, pero que permiten financiar pagos y mantener el nivel de gasto. Estas son las **entradas o gastos compensatorios**, como:

  * La inversi√≥n ($I$)
  * El gasto p√∫blico ($G$)
  * Las transferencias ($Tr$) si estuvieran presentes

* Por otro lado, los **usos del ingreso** que **no se destinan a la compra de bienes y servicios nacionales**, es decir, las **filtraciones**, que incluyen:

  * El ahorro ($S$)
  * Los impuestos ($T$)

El equilibrio se alcanza cuando ambos flujos se igualan:

$\text{Entradas compensatorias} = \text{Filtraciones}$

Este enfoque permite entender el rol activo del Estado en la determinaci√≥n del nivel de actividad econ√≥mica y facilita el an√°lisis de los efectos de la pol√≠tica fiscal. Es particularmente √∫ til para identificar cu√°les son los componentes que dinamizan el gasto y cu√°les act√∫a n como freno al mismo.

En las siguientes secciones, se desarrollar√° la f√≥rmula general del equilibrio y se analizar√° su comportamiento seg√∫n los valores de los distintos par√°metros del modelo.


### üíµ Funci√≥n de ahorro en el modelo keynesiano

El **ahorro (S)** se define como la parte del ingreso disponible que no se destina al consumo. En el modelo keynesiano simple con impuestos auton√≥mos, el ingreso disponible est√° dado por:

$Y_d = Y - T$

Donde:

* $Y$: ingreso total
* $T$: impuestos auton√≥mos

La funci√≥n de consumo es:

$C = C_o + c \cdot (Y - T)$

Donde:

* $C_o$: consumo aut√≥nomo (independiente del ingreso)
* $c$: propensi√≥n marginal a consumir

Entonces, el **ahorro se calcula como**:

$S = Y - T - C$

Sustituyendo la funci√≥n de consumo:

$S = Y - T - (C_o + c(Y - T))$

Desarrollando:

$S = (1 - c)(Y - T) - C_o$

---

### üîé Interpretaci√≥n de los componentes:

* $(1 - c)$: es la **propensi√≥n marginal a ahorrar**. Representa la parte del ingreso disponible que no se consume y se destina al ahorro.
* $(Y - T)$: es el **ingreso disponible**, lo que efectivamente se puede gastar o ahorrar.
* $C_o$: representa un "desahorro" inicial, ya que es el consumo que se realiza incluso sin ingreso. Por eso aparece como una deducci√≥n en la f√≥rmula del ahorro.





Captura de pantalla 2025-06-12 105312.png

```python
# @title
import numpy as np
import matplotlib.pyplot as plt

# Par√°metros del modelo
C0 = 400    # Consumo aut√≥nomo
c = 0.5     # Propensi√≥n marginal a consumir
T0 = 200    # Impuesto aut√≥nomo
I0 = 200    # Inversi√≥n aut√≥noma
G0 = 500    # Gasto p√∫blico

# Eje de ingreso extendido
Y = np.linspace(0, 3000, 300)

# Ingreso disponible y consumo
Yd = Y - T0
C = C0 + c * Yd
S = Yd - C
T = np.full_like(Y, T0)

# Filtraciones = Ahorro + Impuestos
Filtraciones = S + T

# Gastos compensatorios = I + G
Gastos_comp = np.full_like(Y, I0 + G0)

# Ingreso de equilibrio
Y_eq = (C0 - c * T0 + I0 + G0) / (1 - c)

# Multiplicadores
multiplicador_gasto = 1 / (1 - c)
multiplicador_impuestos = -c / (1 - c)

# Gr√°fico
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(Y, Filtraciones, label='Filtraciones (Ahorro + T‚ÇÄ)', color='orange')
ax.plot(Y, Gastos_comp, label='Gastos Compensatorios (I + G)', color='purple')
ax.axvline(x=Y_eq, linestyle=':', color='red', label='Y* (equilibrio)')
ax.text(Y_eq + 20, I0 + G0 + 20, f'Y* = {Y_eq:.1f}', color='red')

# Est√©tica de ejes
ax.set_title('Filtraciones vs. Gastos Compensatorios ‚Äì Equilibrio Keynesiano')
ax.set_xlabel('Renta (Y)')
ax.set_ylabel('Filtraciones y Gastos Compensatorios')
ax.set_xlim(left=0)
ax.set_ylim(-400, max(np.max(Filtraciones), np.max(Gastos_comp)) + 100)
ax.grid(True, linestyle='--', alpha=0.6)

# Mostrar eje vertical con escala, quitar eje horizontal
ax.spines['left'].set_position('zero')
ax.spines['left'].set_color('black')
ax.spines['left'].set_linewidth(1)
ax.spines['bottom'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Leyenda
ax.legend(loc='upper left')
plt.tight_layout()
plt.show()

# Resultados clave
print(f"Ingreso de equilibrio (Y*): {Y_eq}")
print(f"Multiplicador del gasto aut√≥nomo: {multiplicador_gasto}")
print(f"Multiplicador del impuesto aut√≥nomo: {multiplicador_impuestos}")
```

## üí∞ Funci√≥n de Ahorro y su valor en el equilibrio

En el marco del modelo keynesiano, el ahorro se define como la parte del ingreso disponible que no se destina al consumo. Una forma alternativa de expresar la funci√≥n de ahorro es:


S = -C0 + (1 - c)Yd


Donde:

- \( C0 \) es el consumo aut√≥nomo.
- \( c \) es la propensi√≥n marginal a consumir.
- \( Yd = Y - T \) es el ingreso disponible, siendo \( To \) el impuesto aut√≥nomo.

---

### üìå C√°lculo en el equilibrio

Dado el ingreso de equilibrio \( Y* = 2000 \), y con \( To = 200 \), se tiene:

\[
Yd = 2000 - 200 = 1800
\]


S* = -400 + 0,5 x 1800 = 500


Este valor representa el **ahorro en el punto de equilibrio**, y es coherente con el resultado obtenido al igualar las **filtraciones** (ahorro + impuestos) con los **gastos compensatorios** (inversi√≥n + gasto p√∫blico):


S + To = 500 + 200 = 700 = I + G

```python
# @title
##La funci√≥n Ahorro y el nivel de ahorro alcanzado en el equilibrio
import numpy as np

# Par√°metros del modelo
C0 = 400      # Consumo aut√≥nomo
c = 0.5       # Propensi√≥n marginal a consumir
T0 = 200      # Impuesto aut√≥nomo
Y_eq = 2000   # Ingreso de equilibrio previamente calculado

# Ingreso disponible en equilibrio
Yd_eq = Y_eq - T0

# Funci√≥n de ahorro alternativa: S = -C0 + (1 - c) * Yd
S_eq_alt = -C0 + (1 - c) * Yd_eq

# Mostrar resultado
print(f"Ahorro en el equilibrio (por -C0 + (1 - c) * Yd): {S_eq_alt}")
```

```python
# @title
# ‚öôÔ∏è SIMULADOR INTERACTIVO ‚Äì FILTRACIONES Y GASTOS COMPENSATORIOS

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- Widgets de entrada ---
Co_w = widgets.IntSlider(value=400, min=0, max=1000, step=50, description='C‚ÇÄ')
Io_w = widgets.IntSlider(value=200, min=0, max=1000, step=50, description='I‚ÇÄ')
Go_w = widgets.IntSlider(value=500, min=0, max=1000, step=50, description='G‚ÇÄ')
To_w = widgets.IntSlider(value=200, min=0, max=1000, step=50, description='T‚ÇÄ')
c_w = widgets.FloatSlider(value=0.5, min=0.1, max=0.9, step=0.05, description='c')

reset_btn = widgets.Button(description="üîÅ Reiniciar valores")
out = widgets.Output()

# --- Valores iniciales/base ---
Co_base = 400
Io_base = 200
Go_base = 500
To_base = 200
c_base = 0.5

# --- Funci√≥n principal ---
def actualizar_grafico(change=None):
    with out:
        clear_output(wait=True)

        # Valores actuales
        Co = Co_w.value
        Io = Io_w.value
        Go = Go_w.value
        To = To_w.value
        c = c_w.value

        # Ingresos de equilibrio
        Y_base = (Co_base - c_base * To_base + Io_base + Go_base) / (1 - c_base)
        Y_eq = (Co - c * To + Io + Go) / (1 - c)

        Y = np.linspace(0, max(Y_base, Y_eq) + 300, 300)

        # Filtraciones
        F_base = -Co_base + (1 - c_base) * (Y - To_base) + To_base
        F_new = -Co + (1 - c) * (Y - To) + To

        # Gastos compensatorios
        G_base = Io_base + Go_base
        G_new = Io + Go

        # --- Gr√°fico ---
        plt.figure(figsize=(9,6))
        plt.plot(Y, F_base, '--', label='Filtraciones Base', color='blue')
        plt.plot(Y, F_new, '-', label='Filtraciones Nueva', color='orange')
        plt.axhline(y=G_base, linestyle='--', color='gray', label='Gastos Comp. Base')
        plt.axhline(y=G_new, linestyle='-', color='purple', label='Gastos Comp. Nuevo')
        plt.axvline(x=Y_base, linestyle='--', color='black')
        plt.axvline(x=Y_eq, linestyle=':', color='red')
        plt.text(Y_base + 10, G_base + 10, f'Y‚ÇÄ* = {Y_base:.1f}', color='black')
        plt.text(Y_eq + 10, G_new + 10, f'Y* = {Y_eq:.1f}', color='red')

        # Flecha ŒîY
        plt.annotate('', xy=(Y_eq, 0), xytext=(Y_base, 0),
                     arrowprops=dict(arrowstyle='<|-|>', color='crimson', lw=2))
        plt.text((Y_base + Y_eq)/2, 20, f'ŒîY = {Y_eq - Y_base:.1f}',
                 ha='center', va='bottom', fontsize=10, color='crimson')

        # Ordenada al origen de las filtraciones nuevas
        ordenada_origen = -Co + (1 - c) * (-To) + To
        plt.scatter([0], [ordenada_origen], color='brown', zorder=5)
        plt.text(10, ordenada_origen + 20, f'F(0) = {ordenada_origen:.1f}', color='brown')

        plt.xlabel('Ingreso (Y)')
        plt.ylabel('Filtraciones y Gastos Compensatorios')
        plt.title('Modelo de equilibrio: Filtraciones vs. Gastos Compensatorios')
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()

        # --- Tabla de resultados ---
        resultados = {
            "Ingreso base (Y‚ÇÄ*)": Y_base,
            "Ingreso nuevo (Y*)": Y_eq,
            "ŒîY": Y_eq - Y_base
        }
        df_resultado = pd.DataFrame.from_dict(resultados, orient='index', columns=['Valor'])
        display(df_resultado)

# --- Conectar controles ---
for w in [Co_w, Io_w, Go_w, To_w, c_w]:
    w.observe(actualizar_grafico, names='value')

# --- Reinicio ---
def reiniciar(val):
    Co_w.value = Co_base
    Io_w.value = Io_base
    Go_w.value = Go_base
    To_w.value = To_base
    c_w.value = c_base

reset_btn.on_click(reiniciar)

# --- Mostrar controles y resultado ---
display(widgets.HBox([Co_w, Io_w, Go_w, To_w, c_w, reset_btn]))
display(out)
actualizar_grafico()
```

## üìò Incorporaci√≥n de impuestos proporcionales en el modelo keynesiano

Cuando se incorpora una tasa impositiva proporcional al ingreso, representada por t, el ingreso disponible de las familias se reduce. En lugar de recibir el ingreso total Y, las familias disponen de una fracci√≥n menor, equivalente a (1 - t) multiplicado por Y.

Por lo tanto, la funci√≥n de consumo se modifica y pasa a ser:

C = C‚ÇÄ + c(1 - t)Y

La demanda global se construye a partir del consumo, la inversi√≥n y el gasto p√∫blico. Al sustituir la nueva funci√≥n de consumo, la expresi√≥n de la demanda global queda:

Z = C‚ÇÄ + c(1 - t)Y + I‚ÇÄ + G‚ÇÄ

---

### Cambios en la pendiente de la curva Z

Con esta modificaci√≥n, la pendiente de la funci√≥n Z ya no es igual a c, sino que es igual a c por (1 - t). Esto significa que si la tasa impositiva t aumenta, la pendiente de la curva Z disminuye.

Una menor pendiente implica que, ante un mismo incremento del ingreso, el aumento en la demanda es m√°s moderado. Esto provoca una rotaci√≥n de la curva Z hacia abajo, afectando la posici√≥n del equilibrio.

---

### Impacto sobre el ingreso de equilibrio

La nueva expresi√≥n del ingreso de equilibrio se obtiene igualando Z con Y y despejando:

Y = (C‚ÇÄ + I‚ÇÄ + G‚ÇÄ) / [1 - c(1 - t)]

Esta expresi√≥n muestra que el ingreso de equilibrio depende inversamente de la tasa impositiva. A mayor t, menor ingreso de equilibrio.

---

### Interpretaci√≥n

La incorporaci√≥n de impuestos proporcionales introduce un efecto moderador en la din√°mica del modelo. Cuando aumenta el ingreso, tambi√©n lo hacen los impuestos, lo que reduce el ingreso disponible y modera el crecimiento del consumo.

Esto hace que el multiplicador de la demanda global sea m√°s bajo, ya que parte del ingreso generado por un aumento inicial en el gasto se pierde en forma de recaudaci√≥n impositiva.

El resultado es una econom√≠a que responde con menor intensidad a est√≠mulos aut√≥nomos, como una expansi√≥n del gasto p√∫blico, ya que el efecto multiplicador se ve atenuado por la carga tributaria.


## üîÅ Cambios en la pendiente de la curva Z: efectos de c y t

Cuando en el modelo keynesiano simple incorporamos una **tasa impositiva proporcional al ingreso**, representada por \( t \), el ingreso disponible de las familias se reduce a:

$$
Y_d = (1 - t) \cdot Y
$$

Y la funci√≥n de consumo queda expresada como:

$$
C = C_0 + c(1 - t)Y
$$

Por lo tanto, la **Demanda Global** es:

$$
Z = C_0 + c(1 - t)Y + I_0 + G_0
$$

---

### üîÉ ¬øQu√© pasa cuando cambia c o t?

- Un aumento en la **propensi√≥n marginal a consumir** \( c \) **incrementa la pendiente** de la curva \( Z \).
- Un aumento en la **tasa impositiva** \( t \) **reduce la pendiente**, ya que disminuye el ingreso disponible.
- Como resultado, se produce una **rotaci√≥n de la curva Z**, modificando su √°ngulo de inclinaci√≥n.

---

### üìâ Impacto sobre la renta de equilibrio

- Si la pendiente de \( Z \) se vuelve m√°s pronunciada, el ingreso de equilibrio \( Y^* \) aumenta.
- Si la pendiente disminuye (por ejemplo, por una suba en \( t \)), el ingreso de equilibrio cae.
- La ordenada al origen \( A \) tambi√©n puede cambiar seg√∫n los valores de \( c \), \( t \), y los componentes aut√≥nomos.


```python
# ‚öôÔ∏è Simulador interactivo con impuestos proporcionales y multiplicadores

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- Deslizadores de entrada ---
C0_w = widgets.IntSlider(value=400, min=0, max=1000, step=50, description='C‚ÇÄ')
I0_w = widgets.IntSlider(value=200, min=0, max=1000, step=50, description='I‚ÇÄ')
G0_w = widgets.IntSlider(value=500, min=0, max=1000, step=50, description='G‚ÇÄ')
T0_w = widgets.IntSlider(value=200, min=0, max=1000, step=50, description='T‚ÇÄ')
c_w = widgets.FloatSlider(value=0.75, min=0.1, max=0.95, step=0.05, description='c')
t_w = widgets.FloatSlider(value=0.2, min=0.0, max=0.5, step=0.05, description='t')

reset_btn = widgets.Button(description="üîÅ Reiniciar valores")
out = widgets.Output()

# --- Valores base ---
C0_base = 400
I0_base = 200
G0_base = 500
T0_base = 200
c_base = 0.5
t_base = 0.0

# --- Funci√≥n principal ---
def actualizar_modelo(change=None):
    with out:
        clear_output(wait=True)

        # Valores actuales desde los widgets
        C0 = C0_w.value
        I0 = I0_w.value
        G0 = G0_w.value
        T0 = T0_w.value
        c = c_w.value
        t = t_w.value

        # --- Equilibrio base (sin impuestos proporcionales, c_base) ---
        A_base = C0_base + I0_base + G0_base - c_base * T0_base
        k_base = 1 / (1 - c_base)
        Y_base = A_base * k_base

        # --- Equilibrio con t > 0 y nueva c ---
        A = C0 + I0 + G0 - c * T0
        k = 1 / (1 - c * (1 - t))
        Y_eq = A * k

        # --- Multiplicadores ---
        k_T = -c / (1 - c * (1 - t))
        k_G = k

        # --- Eje ingreso ---
        Y = np.linspace(0, max(Y_base, Y_eq) + 300, 300)
        Z_base = A_base + c_base * Y
        Z = A + c * (1 - t) * Y

        # --- Gr√°fico ---
        plt.figure(figsize=(9,6))
        plt.plot(Y, Y, '--', color='gray', label='Oferta global (Y)')
        plt.plot(Y, Z_base, label='Demanda global base Z‚ÇÄ', color='blue')
        plt.plot(Y, Z, label='Demanda global nueva Z', color='green')
        plt.scatter([Y_base], [Y_base], color='blue')
        plt.text(Y_base + 10, Y_base - 30, f'Y‚ÇÄ* = {Y_base:.1f}', color='blue')
        plt.scatter([Y_eq], [Y_eq], color='green')
        plt.text(Y_eq + 10, Y_eq - 30, f'Y* = {Y_eq:.1f}', color='green')

        # Flecha ŒîY
        plt.annotate('', xy=(Y_eq, 0), xytext=(Y_base, 0),
                     arrowprops=dict(arrowstyle='<|-|>', color='crimson', lw=2))
        plt.text((Y_base + Y_eq)/2, 20, f'ŒîY = {Y_eq - Y_base:.1f}',
                 ha='center', va='bottom', fontsize=10, color='crimson')

        # Ordenada del nuevo Z
        plt.scatter([0], [A], color='purple')
        plt.text(10, A + 10, f'A = {A:.1f}', color='purple')

        plt.xlabel('Ingreso (Y)')
        plt.ylabel('Demanda Global (Z)')
        plt.title('Modelo Keynesiano con impuestos proporcionales')
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()

        # --- Tabla de resultados ---
        resultados = {
            "Ingreso base (Y‚ÇÄ*)": Y_base,
            "Ingreso nuevo (Y*)": Y_eq,
            "ŒîY": Y_eq - Y_base,
            "Multiplicador del Gasto (k)": k_G,
            "Multiplicador de los Impuestos (kT)": k_T
        }
        df_resultado = pd.DataFrame.from_dict(resultados, orient='index', columns=['Valor'])
        display(df_resultado)

# --- Conectar los sliders ---
for w in [C0_w, I0_w, G0_w, T0_w, c_w, t_w]:
    w.observe(actualizar_modelo, names='value')

# --- Funci√≥n de reinicio ---
def reiniciar(val):
    C0_w.value = C0_base
    I0_w.value = I0_base
    G0_w.value = G0_base
    T0_w.value = T0_base
    c_w.value = 0.75  # valor modificado
    t_w.value = 0.2   # impuesto proporcional activado

reset_btn.on_click(reiniciar)

# --- Mostrar interfaz y resultado ---
display(widgets.HBox([C0_w, I0_w, G0_w, T0_w, c_w, t_w, reset_btn]))
display(out)
actualizar_modelo()
```
